```python
'''Using imblearn library to use oversampling method (SMOTE) to handle class imbalances'''
```

```python
import basic_functions

df = basic_functions.read_csv('flight.csv')
df = basic_functions.drop(df)
df.dropna(inplace=True)
```

```python
delay_count = df['ARR_DEL15'].value_counts()
delay_count.plot(kind = 'bar')
```

```python
import numpy as np
import matplotlib.pyplot as plt
X = df.values
y = df['ARR_DEL15']
def plot_2d_space(X, y, label='Classes'):   
    colors = ['#1F77B4', '#FF7F0E']
    markers = ['o', 's']
    for l, c, m in zip(np.unique(y), colors, markers):
        plt.scatter(
            X[y==l, 0],
            X[y==l, 1],
            c=c, label=l, marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()
    
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X = pca.fit_transform(X)

plot_2d_space(X, y, 'Imbalanced Flight Delay Data (2 PCA components)')
```
